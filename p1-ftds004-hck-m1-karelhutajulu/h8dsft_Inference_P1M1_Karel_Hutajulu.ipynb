## 2. DATA LOADING ðŸ’½

```python
df = pd.read_csv('/Users/karelhutajulu/Downloads/h8dsft_P1M1_Karel_Hutajulu.csv')
```

```python
#Display top 10 rows of dataframe df
df.head(10)
```

```python
inference = df.sample(3)
```

```python
### Import Models and applying prediction
```

```python

# Load the saved model using joblib
loaded_model4 = joblib.load('encoder.pkl')
loaded_model2 = joblib.load('standard_scaler.pkl')
loaded_model3 = joblib.load('minmax_scaler.pkl')
loaded_model1 = joblib.load('rf_model_best.pkl')
```

```python
# Apply the scaling and encoding models to the inference DataFrame

inference[['sex', 'marital_status','education_level','pay_0', 'pay_2', 'pay_3', 'pay_4', 'pay_5', 'pay_6']] = loaded_model4.transform(inference[['sex', 'marital_status','education_level','pay_0', 'pay_2', 'pay_3', 'pay_4', 'pay_5', 'pay_6']])
inference['age'] = loaded_model2.transform(inference[['age']])
inference[['limit_balance', 'bill_amt_1', 'pay_amt_1', 'pay_amt_2', 'pay_amt_3', 'pay_amt_4', 'pay_amt_5', 'pay_amt_6']] = loaded_model3.transform(inference[['limit_balance', 'bill_amt_1', 'pay_amt_1', 'pay_amt_2', 'pay_amt_3', 'pay_amt_4', 'pay_amt_5', 'pay_amt_6']])
```

```python

# Apply the prediction model to the new DataFrame
predictions1 = loaded_model1.predict(inference)



# Add the predictions as a new column to the DataFrame
inference['target_prediction1'] = predictions1
```
